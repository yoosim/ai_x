```python
# 맨처음 import로 해야할 것 

import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential   # 모델 객체 생성
from tensorflow.keras.layers import Input, Dense # layer층을 쌓을 때 필요
```

```python
X_data = np.array([[0,0],
                   [0,1],
                   [1,0],
                   [1,1]])

y_data = np.array([[0],[1],[1], [0]])
```

```python
# 3. 모델 구성
# layer 층 x_data 2개 -> 10개 -> 20개 -> 10개 -> 1개
model = Sequential()  # 모델 구성
# model.add(Input(shape=(2,)))        # 입력층
model.add(Dense(units=10,            # 출력
                input_dim=2,          # 입력 input_dim=2 , Input(shape=(2,))이걸 썼으면 input_dim안 써도 됨
                activation='relu'))   # 활성화 함수
model.add(Dense(units=20,activation='relu'))
model.add(Dense(units=10,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))   
# 이진분류에서의 출력층 활성화 함수는 sigmoid
# 다중분류에서의 출력층 활성화 함수 softmax (출력합을 1로 만드는것)
model.summary()
```

```python
# 4. 모델 학습과정 설정
model.compile(loss='binary_crossentropy',   # 다중분류면 crossentropy
               optimizer='adam',
               metrics=['binary_accuracy']) 
```

```python
# 5. 학습

hist= model.fit(X_data, y_data,   # 훈련 데이터
                epochs=600,       # 학습 횟수
#                 batch_size=4    # 메모리에 따라서 설정
                verbose=2         # 학습로그 출력 0-출력 안됨, 1-간단, 2-상세출력
               )
```

```python
# 5. 학습과정 살펴보기
hist.history.keys()   # compile할 때 metrics에 binary_accuracy를 넣어서 loss와 binary_accuracy가 나타남
```

```python
import matplotlib.pyplot as plt
fig, loss_ax = plt.subplots(figsize=(10,5))
loss_ax.plot(hist.history['loss'], 'r', label='loss')
loss_ax.set_xlabel('epochs')
loss_ax.set_ylabel('loss')
loss_ax.legend(loc='center right')
acc_ax = loss_ax.twinx()    #loss_ax와 x축을 공유하는 acc_ax 생성
acc_ax.plot(hist.history['binary_accuracy'], 'b', label='binary_accuracy')
acc_ax.set_ylabel('binary_accuracy')
acc_ax.legend(bbox_to_anchor=(0.999, 0.48))
```

```python
# 6. 학습평가하기
score = model.evaluate(X_data, y_data)
print('loss :', score[0])
print('binary_accuracy :', score[1])
```

```python
# 7. 모델 사용
model.predict(np.array([[0,1]])).round()   #2차원으로 넣어야 함  / 반올림하면 1로 나타남 /바이너리 회귀분석
```
