```python
# 말뭉치에서 단어추출 (특수문자 제외하고 2글자 이상인 단어) 
from nltk.tokenize import RegexpTokenizer
from nltk.tag import pos_tag
word = RegexpTokenizer('\w{2,}').tokenize(emma) # \w = [a-zA-Z0-9]
word
# 명사만 추출
emma_tags = pos_tag(word)
nous_list = [token for token, tag in emma_tags if tag in ('NN','NNS')]
# nous_list[:5]
print('추출된 명사 갯수 :',len(nous_list), '\t단어 종류 수 :',len(set(nous_list)))
print('\nnoun_list :', nous_list[:10])
emma_noun = ' '.join(nous_list)
print('\n워드 클라우드에 필요한 내용 :',emma_noun[:100])
```

```python
#웹크롤링
import requests
from bs4 import BeautifulSoup
rss_url = 'https://fs.jtbc.co.kr/RSS/economy.xml'
jtbc_response = requests.get(rss_url)
soup = BeautifulSoup(jtbc_response.content, "xml")
link_el = soup.select('item link')
link_list = [ el.text for el in link_el]
len(link_list)
```

```python
from konlpy.tag import Komoran
komoran = Komoran()
news = []
for link in link_list:
    response = requests.get(link)
#     print(response.status_code)
    news_soup = BeautifulSoup(response.content, "html.parser")
    title = news_soup.select('title')
    print(title)
```

```python
from konlpy.tag import Komoran
komoran = Komoran()
news = []
for link in link_list:
    response = requests.get(link, timeout=10)
#     print(response.status_code)
    news_soup = BeautifulSoup(response.content, "html.parser")
    title = news_soup.select('title')[0].text
    description = str(news_soup.select_one('meta[name="description"]'))
    noun_list = komoran.nouns(title + ' ' + description)
    불용어 = {'뉴스','앵커'}
    noun_list = [word for word in noun_list if word not in 불용어]
#     print(noun_list)
    news.append(noun_list)
print(news)
```

```python
# rss의 title과 description을 가지와 명사 추출 후 워드임베딩(단어간 거리 계산 / 백터 계산) = 특정 단어의 유사도 출력
import requests
from bs4 import BeautifulSoup
rss_url = 'https://fs.jtbc.co.kr/RSS/economy.xml'
jtbc_response = requests.get(rss_url)
soup = BeautifulSoup(jtbc_response.content, "xml")

news = []
item_elems =  soup.find_all('item')

for item_elem in item_elems:
    title = item_elem.find('title').text
    description = item_elem.find('description').text
#     print(title + ' ' + description )
    article = title + ' ' + description.replace('[앵커]',' ')
    noun_lis = komoran.nouns(article) 
    # 보통명사(NNG),고유명사(NNP)
    noun_list = [word for word, tag in komoran.pos(article) if tag in ('NNG','NNP')]
#     print(noun_list1)
#     print(noun_list)
    news.append(noun_list)
print(news[:3])
```

```python
# 워드 임베딩 학습 (단어간 거리 계산)
# pip install gensim 
# 텍스트를 넘파이 숫자로 변환
from gensim.models import Word2Vec
model = Word2Vec(news,        # 학습데이터 2차원
                 window=10,   # 각 단어의 좌우10개 단어를 학습 컨텐츠로 사용
                 min_count=2, # 학습데이터에서의 2회 이상 반복된 단어만 학습
                 workers=-1,  # 병렬처리 core수. -1:모든 가용 코어 사용
                )
```
