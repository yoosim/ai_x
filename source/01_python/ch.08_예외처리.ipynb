## 2.2 예외를 지정한 처리

```
try: 
    예외가 발생할 수도 있는 명령어들
except 예외타입1:
    하당 예외타입이 발생할 경우 실행할 문장
except 예외타입2:
    하당 예외타입이 발생할 경우 실행할 문장    
```

분할해서 예외타입에 따라 예외 처리가 가능함.

````python
# ZeroDivisionError : division by zero 예외객체 : 예외 메세지
# ValueError : invalid literal for int() with base 10: 'a'


while True:
    try:
        x = int(input('정수를 입력하세요 :'))
        result = 100/x
        result = f'{100/x:.0f}' if round(result)==result else f'{100/x:.2f}'
        print(result)
        break
        
    except ValueError:
        print('★정수를 입력하세요!!!!!!!★')
    except ZeroDivisionError:
        print('0으로 나눌 수 없어요.')
    except Exception:  #Exception()은 모든 예외처리의 상위버전  , ()에는 타입을 써야 함
        print('그 외의 다른 예외, 다시 시도하세요.')
        
        
````

````python
# except 순서를 바꿔서 상위타입이 제일 위로 올리면, 하위 에러보다 상위 에러 멘트가 나옴.
# except절이 많은 경우 하위 클래스(자식클래스)가 위에 나와야 함.

while True:
    try:
        x = int(input('정수를 입력하세요 :'))
        result = 100/x
        result = f'{100/x:.0f}' if round(result)==result else f'{100/x:.2f}'
        print(result)
        break
        
    except Exception:  #Exception()은 모든 예외처리의 상위버전  , ()에는 타입을 써야 함
        print('그 외의 다른 예외, 다시 시도하세요.')        
    except ValueError:
        print('★정수를 입력하세요!!!!!!!★')
    except ZeroDivisionError:
        print('0으로 나눌 수 없어요.')
````

````python
# 여러 예외처리 문제를 합해서 정리 할 수 있음

while True:
    try:
        x = int(input('정수를 입력하세요 :'))
        result = 100/x
        result = f'{100/x:.0f}' if round(result)==result else f'{100/x:.2f}'
        print(result)
        break
             
    except (ZeroDivisionError, ValueError):
        print('★잘 입력하세요')
    except Exception:  #Exception()은 모든 예외처리의 상위버전  , ()에는 타입을 써야 함
        print('그 외의 다른 예외, 다시 시도하세요.')  
````

## 2.3 예외메세지(예외인수)


````python
# ZeroDivisionError : division by zero 예외객체 : 예외 메세지
# ValueError : invalid literal for int() with base 10: 'a'

while True:
    try:
        x = int(input('정수를 입력하세요 :'))
        result = 100/x
        result = f'{100/x:.0f}' if round(result)==result else f'{100/x:.2f}'
        print(result)
        break
             
    except (ZeroDivisionError, ValueError) as e:
        print(e)                                  #e.__str__() : 예외메세지를 리턴해준다.
        print(type(e))
        print('예외메세지 :', e.args)               #e.args는 튜플
        print('예외메세지 :', e.args[0])
    except Exception:                             #Exception()은 모든 예외처리의 상위버전  , ()에는 타입을 써야 함
        print('그 외의 다른 예외, 다시 시도하세요.')  
````

### try문법

```
try: 
    예외가 발생할 수도 있는 구문
except [예외타입 as e]:
    예외가 발생되면 수행할 구문
else:
    try절을 수행하다 예외가 발생 안 될면 else절을 한번 수행
finally:
    예외가 발생 안 하면 try-else-finally 실행
    예외가 발생하면 try수행중 except-finally절 수행
```

````python
try:
    f = open('data/ch.08.txt', 'r')
except FileNotFoundError as e:
    print(e)
    print('해당 파일이 없습니다.')
else:
    data = f.read()
    print(data)
finally:
    f.close()   
````

````python
#else절이 빠진 try절 최종 정리
try:
    f = open('data/ch.08.txt', 'r')
    data = f.read()
    print(data)
except FileNotFoundError as e:
    print(e)
finally:
    f.close()   
````

````python
# 사용자 정의 예외 : Exception 클래스 또는 그 하위 클래스를 상속받아 구현
# 중첩루프를 빠져나올때.. 
class LangthZeroError(Exception): #Exception안에 __init__(self, message), self.messgae = message 있음
    '길이가 0일 때 발생할 예외'
    pass
#     def __init__(self, message)
#             self.massage = message
````

````python
#매개변수가 있도록 만들었음

def insert(*data): #튜플 매개변수
    if len(data)==0:
        raise LangthZeroError("매개변수가 0인 예외입니다.")  #사용자 정의 예외 발생
    for item in data:
        print(item, end=' ') #한칸씩 띄우고 출력
    print('등을 입력하셨습니다.')
````

````python
data = []
try:
    insert(*data) #언패킹
except LangthZeroError as e:
    print(e)
finally:
    print('무조건 실행하는 부분 DONE')
````

# 4절. 추상클래스와 추상메서드
- 추상클래스 : 객체를 생성할 수 없는 클래스 (상속받을 클래스의 포맷만 제공)
- 추상메소드 : 호출할 수 없는 메소드

````python
# Shape 클래스 : 객체를 만드는게 목적이 아니고, 자식클래스(서브클래스)를 잘 만들게 목적
# Shape 클래스를 상속받은 클래스들의 표준 제시
# 생성자 함수를 못 쓰게 만든다.
# NotImplementedError 추상클래스를 만들때 쓰는것 
class Shape:
    def __init__(self):
        raise NotImplementedError('추상 클래스 역할')
    def calc_area(self):
        raise NotImplementedError('추상 메소드 역할')
        
        
    
````

````python
#s = Shape() #상속을 받을 수 없어서 오류가 남 
class Circle(Shape):
    def __init__(self, r=3):   #여기서 r은 반지름
        self.r = r
    def calc_area(self):
        return 3.14*self.r*self.r
c = Circle(3)
c.calc_area()
````
