```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina' 
#ÌïúÍ∏ÄÏÑ§Ï†ï #rcÎäî plt Í∏∞Î≥∏ ÏÜçÏÑ±Îì§ÏùÑ ÏÑ§Ï†ïÌïòÎäî Ìï®Ïàò
plt.rc('font',family='Malgun Gothic') #ÏúàÎèÑÏö∞Ï¶à
plt.rc('axes', unicode_minus=False) #Ï∂ïÏÑ§Ï†ï (Ï∂ïÏùò - Íπ®Ïßê Î∞©ÏßÄ)
```

```python
# Î∂ÑÎ•òÎ•º ÏúÑÌïú Í∞ÄÏÉÅÏùò Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
from sklearn.datasets import make_classification 
X, y = make_classification(n_samples=20,             # Îç∞Ïù¥ÌÑ∞ Í∞ØÏàò (Í∏∞Î≥∏Í∞í-100)
                           n_features=2 ,            # ÎèÖÎ¶ΩÎ≥ÄÏàò Í∞ØÏàò (Í∏∞Î≥∏Í∞í-20) 
                           n_informative=2,          # Ï¢ÖÏÜçÎ≥ÄÏàòÏóê ÏòÅÌñ•ÏùÑ ÎØ∏ÏπòÎäî ÎèÖÎ¶ΩÎ≥ÄÏàò Í∞ØÏàò 
                           n_redundant=0,            # Ï¢ÖÏÜçÎ≥ÄÏàòÏóê ÏòÅÌñ•ÏùÑ ÎØ∏ÏπòÏßÄÏïäÎäî ÎèÖÎ¶ΩÎ≥ÄÏàò Í∞ØÏàò
                           n_classes=2 ,             # Ï¢ÖÏÜçÎ≥ÄÏàòÏùò Í∑∏Î£π Ïàò(0,1)
                           n_clusters_per_class=1,   # Í∞Å Ï¢ÖÏÜçÎ≥ÄÏàò Í∑∏Î£πÎ≥Ñ ÏÑúÎ∏åÍ∑∏Î£π Ïàò (1Í∞úÎùºÎäîÍ±¥ ÏÑúÎ∏åÍ∑∏Î£π ÏóÜÏùå)
                           random_state=123)
import numpy as np
np.c_[X,y]
```

```python
plt.scatter(x = X[:,0],
            y = X[:,1],
            c=y, s=100, edgecolors='k', lw=3)
plt.show()
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=2,       # Íµ∞Ïùò Í∞ØÏàò
               init="k-means++",   # k-means++(Ï¥àÍ∏∞Ï§ëÏã¨Ï†êÏùÑ ÏÑúÎ°ú Î©ÄÎ¶¨ Ï¥àÍ∏∞Ìôî)/ random(Ï¥àÍ∏∞Ï§ëÏã¨Ï†êÎûúÎç§Ï¥àÍ∏∞Ìôî)
               n_init=10,          # Í∞Å Îã§Î•∏ Ï¥àÍ∏∞Í∞íÏùÑ 10Î≤à Ïã§ÌñâÌï¥Î≥¥Í≥† Í∞ÄÏû• Ï¢ãÏùÄ Î™®Îç∏ ÏÑ†ÌÉù
               max_iter=300)       # Í∞Å Ïã§ÌñâÏóêÏÑú 300Î≤à Î∞òÎ≥µ(Ï§ëÏã¨Ï†ê ÏòÆÍ∏∞Í∏∞)

model.fit(X)              
```

```python
c0, c1 = model.cluster_centers_   # Î™®Îç∏Ïù¥ Ï†ïÌïú ÎëêÍ∞úÏùò Ï§ëÏã¨Ï†ê
print(c0)
print(c1)
```

```python
print('modelLabel :',model.labels_)
pred = model.predict(X)
print("  Ïòà Ï∏° Í∞í  :",pred)
print('  Ïã§ Ï†ú Í∞í  :', y)
```

```python
plt.scatter(x = X[:,0],
            y = X[:,1],
            c=pred, s=100, edgecolors='k', lw=3)
plt.scatter(x = c0[0], y = c0[1], s=200, marker='^', lw=5, c='g')
plt.scatter(x = c1[0], y = c1[1], s=200, marker='^', lw=5, c='g' )
plt.show()
```

```python
np.random.seed(0)
# ÌèâÍ∑†:-10, ÌëúÏ§ÄÌé∏Ï∞®:2Ïù∏ 10Ìñâ 2Ïó¥ Îç∞Ïù¥ÌÑ∞ : group0
group0 = np.random.normal(-10,2,(10,2))
y = np.full(group0.shape[0],0)
group0 = np.c_[group0,y]
group0[:2]
```

```python
# ÌèâÍ∑†:10, ÌëúÏ§ÄÌé∏Ï∞®:2Ïù∏ 10Ìñâ 2Ïó¥ Îç∞Ïù¥ÌÑ∞ : group1
group1 = np.random.normal(10,2,(10,2))
y = np.full(group1.shape[0],1)
group1 = np.c_[group1,y]
group1[:2]
```

```python
# ÌèâÍ∑†:0, ÌëúÏ§ÄÌé∏Ï∞®:5Ïù∏ 100Ìñâ 2Ïó¥ Îç∞Ïù¥ÌÑ∞ : group2
group2 = np.random.normal(0,5,(100,2))
y = np.full(group2.shape[0],2)
group2 = np.c_[group2,y]
group2[:2]
```

```python
data = np.r_[group0, group1, group2]
data.shape, data[::10]
```

```python
plt.scatter(x=data[:,0], y=data[:,1], c=data[:,2], s=100, edgecolors='k')
plt.show()
```

```python
model = KMeans(n_clusters=3,
               init="k-means++",
               n_init="auto",
               max_iter=300
               )
model.fit(data[:,:2])
```

```python
# Íµ∞Ïùò Ï§ëÏã¨Ï†ê
c = model.cluster_centers_
c
```

```python
# Ïñ¥ÎñªÍ≤å ÎÇòÎà¥ÎäîÏßÄ Ï≤¥ÌÅ¨ Î∞©Î≤ï (Îëò Îã§ ÎòëÍ∞ôÏùå)
pred = model.labels_
pred = model.predict(data[:,:2]) # Î™®Îç∏Ïù¥ ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅÌïú Í≤∞Í≥º (Íµ∞ÏßëÌôî)
pred
```

```python
plt.scatter(x=data[:,0], y=data[:,1], c=pred, s=100, edgecolors='k')
plt.scatter(x=c[:,0],y=c[:,1], c='r', marker='x', s=200, lw=3)
plt.show()
```

# 3Ï†à. Hierarchical ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ(Í≥ÑÏ∏µÏ†Å Íµ∞Ïßë)
- Íµ∞ÎÅºÎ¶¨ Î¨∂ÏúºÎ©¥ÏÑú ÏµúÏ¢ÖÏ†ÅÏúºÎ°ú ÌïòÎÇòÏùò Íµ∞ÏßëÏù¥ Îê† ÎïåÍπåÏßÄ Î¨∂Îäî ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ
- Íµ∞ÏßëÏùò ÏàòÎ•º ÎØ∏Î¶¨ Ï†ïÌï¥Ï£ºÏßÄ ÏïäÏùå
- Íµ∞ÏßëÍ∞ÑÏùò Í±∞Î¶¨ Í∏∞Î∞òÏúºÎ°ú ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ 

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head(2)
```

```python
# Í≥ÑÏ∏µÏ†Å Íµ∞Ïßë Ï†ÅÏö©ÏùÑ ÏúÑÌï¥ ÎùºÎ≤® Ïù∏ÏΩîÎî©
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
iris['species'] = le.fit_transform(iris.species)
```

```python
iris[::50]
```

```python
from scipy.cluster.hierarchy import linkage
# method : Í±∞Î¶¨Í≥ÑÏÇ∞ÏùÑ Ïñ¥ÎñªÍ≤å Ìï† ÏßÄ
# method='complete' : Îëê ÌÅ¥Îü¨Ïä§ÌÑ∞ÏóêÏÑú Í∞ÄÏû• Î®º Í±∞Î¶¨Î•º Ïù¥Ïö©
# method='single' : Îëê ÌÅ¥Îü¨Ïä§ÌÑ∞ÏóêÏÑú Í∞ÄÏû• Î®º Í±∞Î¶¨Î•º Ïù¥Ïö©
# method='average' : Îëê ÌÅ¥Îü¨Ïä§ÌÑ∞ÎÇ¥Ïùò Í∞Å Ï†êÏóêÏÑú Îã§Î•∏ ÌÅ¥Îü¨Ïä§ÌÑ∞Ïùò Î™®Îì† Ï†êÏùò ÌèâÍ∑†Í±∞Î¶¨ Ïù¥Ïö© 
cluster_model = linkage(iris, method='complete') 
```

```python
from scipy.cluster.hierarchy import dendrogram
dendrogram(cluster_model, labels=iris.index)
# plt.savefig('data/ch02_iris.png', dpi=300, bbox_inches="tight")
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster # Íµ∞ÏßëÌôî
fcluster(cluster_model,
         t=4,  # dendrogramÏùò yÏ∂ïÏùò Í∞í
         criterion='distance' # Íµ∞ÏßëÌôî Í∏∞Ï§Ä
        )
# criterion="distance"Ïùº Îïå, tÍ∞íÏùÄ dendrogramÏùò yÏ∂ï Í∞í
```

```python
fcluster(cluster_model, 
        t=3, 
        criterion='maxclust')
# criterion='maxclust' Ïùº Îïå, tÍ∞íÏùÄ ÌÅ¥Îü¨Ïä§ÌÑ∞Ïùò Í∞ØÏàò(Íµ∞ÏßëÌôîÎêú Í∑∏Î£π Í∞ØÏàò)
```

```python
fcluster(cluster_model, 2, criterion='distance')
```

# 4Ï†à. DBSCAN ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ
- Ïó∞Í≤∞ Í∏∞Î∞ò(Î∞ÄÎèÑÍ∏∞Î∞ò) Íµ∞ÏßëÌôî ÏïåÍ≥†Î¶¨Ï¶ò
- Îã®Ï†ê : ÌÅ¥Îü¨Ïä§ÌÑ∞Ïùò Î∞ÄÎèÑÍ∞Ä ÏùºÏ†ïÌïòÏßÄ ÏïäÏúºÎ©¥ ÏÑ±Îä•Ïù¥ ÎÇÆÏïÑÏßê 

```python
X1 = np.random.rand(200)*3 # 0<=X1<2Ïùò Ïã§Ïàò ÎÇúÏàò 
noise = np.random.normal(0,0.1,200)
Y1 = np.sin(X1) + noise
group1 = np.c_[X1,Y1, np.full(200,0)]
X2 = X1 + 1.5
Y2 = np.cos(X2) + noise
# plt.scatter(X2, Y2)
group2 = np.c_[X2, Y2, np.full(200,1)]
data = np.r_[group1,group2]

plt.scatter(data[:,0], data[:,1], c = data[:,2], s = 100, edgecolors='k', lw=2)
```

```python
data[:,:6]
```

```python
from sklearn.cluster import DBSCAN
# Î∞òÍ≤Ω 0.3 Ïù¥ÎÇ¥Ïóê ÏµúÏÜå 10Í∞úÎäî ÏûàÏñ¥Ïïº Íµ∞Ïßë ÏãúÏûë
db = DBSCAN(eps=0.3, min_samples=10)
db.fit(data[:,:2])
```

```python
labels = db.labels_ # Î™®Îç∏Ïù¥ Íµ∞ÏßëÌôîÌïú Í≤∞Í≥º (-1ÏùÄ Ïù¥ÏÉÅÏπò Ïù∏Ïãù)
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0) # ÌÅ¥Îü¨Ïä§ÌÑ∞ Ïàò
n_noise_ = list(labels).count(-1)
print(f'Ïù¥ÏÉÅÏπòÎ•º Ï†úÏô∏Ìïú ÌÅ¥Îü¨Ïä§ÌÑ∞ Ïàò:{n_clusters_}, Ïù¥ÏÉÅÏπòÍ∞ØÏàò:{n_noise_}Í∞ú')
```

```python

plt.scatter(data[:,0], data[:,1], c = labels, s = 100, edgecolors='k', lw=2)
```

## 5.1 metrics Ìï®Ïàò Ïù¥Ïö© 
- ÌÅ¥Îü¨Ïä§ÌÑ∞Ïùò ÏÜåÏÜçÍ≥º Í∞ØÏàòÎ•º Î™®Îëê ÏïåÍ≥† ÏûàÏùÑ Í≤ΩÏö∞ ÌèâÍ∞Ä 
        

```python
iris.sample()
```

```python
iris_X = iris.iloc[:,:-1]
iris_y = iris.iloc[:,-1]
iris_X.shape, iris_y.shape
```

```python
iris_model3 = KMeans(n_clusters=3,n_init=10, random_state=1)
iris_model3.fit(iris_X)
```

```python
pred = iris_model3.labels_
pred = iris_model3.predict(iris_X) # ÏúÑ ÏïÑÎûò ÎèôÏùºÌïú Î∞©Î≤ï 
```

```python
# pred Ï°∞Ï†ïÏùÑ ÏúÑÌï¥ : np.choose(Î∞∞Ïó¥ÏãúÌÄÄÏä§, Î∞∞Ïó¥)

np.choose([3,2,2,3],[0, 10, 20, 30])
```

```python
pred3 = np.choose(iris_model3.labels_,[1,0,2])
pred3 # Ï°∞Ï†ïÎêú ÏòàÏ∏°Í∞í
```

```python
# 3Í∞ú clusterÎ°ú Íµ∞ÏßëÌôîÌïú ÏûêÎ£å 
print('Ïã§Ï†úÍ∞í :\n', iris_y[::50])
print('ÏòàÏ∏°Í∞í :\n', pred3[::50])
```

```python
iris_model2 = KMeans(n_clusters=2, n_init=10, random_state=1)
iris_model2.fit(iris_X)
```

```python
pred2 = iris_model2.labels_
```

```python
# 2Í∞ú clusterÎ°ú Íµ∞ÏßëÌôîÌïú ÏûêÎ£å 
iris_y2 = np.array([0]*50 + [1]*100)
print('Ïã§Ï†úÍ∞í :\n',iris_y2)
print('ÏòàÏ∏°Í∞í :\n', pred2)
```

```python
# ÌÅ¥Îü¨Ïä§ÌÑ∞Í∞Ä 3Í∞úÏù∏ Í≤ΩÏö∞ ÌèâÍ∞Ä : adjusted_rand_score
from sklearn.metrics import adjusted_rand_score
adjusted_rand_score(labels_true=iris_y,   # Ïã§Ï†úÍ∞í
                    labels_pred=pred3)    # ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ ÏòàÏ∏°
```

```python
# ÌÅ¥Îü¨Ïä§ÌÑ∞Í∞Ä 2Í∞úÏù∏ Í≤ΩÏö∞ ÌèâÍ∞Ä : adjusted_rand_score
adjusted_rand_score(labels_true=iris_y2, # Ïã§Ï†úÍ∞í
                    labels_pred=pred2)   # ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ ÏòàÏ∏°
```

## 5.2 Ïã§Î£®Ïó£ Í≥ÑÏàò
- ÌÅ¥Îü¨Ïä§ÌÑ∞ Í∞úÏàò, ÏÜåÏÜçÏùÑ Î™®Îëê Î™®Î•¥Í≥† ÏûàÏùÑ Í≤ΩÏö∞ ÌèâÍ∞Ä

```python
X1 = np.random.rand(200)*3 # 0<=X1<2Ïùò Ïã§Ïàò ÎÇúÏàò 
noise = np.random.normal(0,0.1,200)
Y1 = np.sin(X1) + noise
group1 = np.c_[X1,Y1, np.full(200,0)]
X2 = X1 + 1.5
Y2 = np.cos(X2) + noise
# plt.scatter(X2, Y2)
group2 = np.c_[X2, Y2, np.full(200,1)]
data = np.r_[group1,group2]

plt.scatter(data[:,0], data[:,1], c = data[:,2], s = 100, edgecolors='k', lw=2)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
range_n_clusters = [2,3,4,5,6,7,8,9,10]
for n_clusters in range_n_clusters:
    model = KMeans(n_clusters=n_clusters,
                   n_init=10,
                   random_state=2)
    model.fit(data[:,:-1])
    score = silhouette_score(X=data[:,:-1],
                             labels=model.labels_)
    print("ÌÅ¥Îü¨Ïä§Ìä∏ Ïàò : {}, Ïã§Î£®Ïó£ Í≥ÑÏàò : {}".format(n_clusters,score))
# Ïã§Î£®Ïó£ Í≥ÑÏàòÍ∞Ä Í∞ÄÏû• ÎÜíÏùÄ 6Í∞úÎ°ú ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ Ìï† ÏòàÏ†ï
```

```python
model = KMeans(n_clusters=6, n_init=10, init="k-means++")
model.fit(data[:,:-1])
centers = model.cluster_centers_
colors = ['r','g','b','c','m','k'] # 4,2,3 vs 0,1,5
for i in range(len(colors)):
    plt.scatter(data[model.labels_==i,0], data[model.labels_==i,1], c=colors[i])
```
