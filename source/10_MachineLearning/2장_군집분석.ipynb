# 2절. K-Means 클러스터링 
- 클러스터(cluster) : 독립변수의 특성이 유사한 데이터의 그룹 
- 클러스터링(clustering) : 주어진 데이터를 여러개의 클러스터로 구분하는 것
- 가장 단순하고 빠른 클러스터링 알고리즘 / 거리 중심
- 클러스터의 중심점을 찾아가는 알고리즘
- 목적함수 값이 최소화될 때까지 클러스터의 중심(centroid) 𝜇k와 각 데이터가 소속될 클러스터를 반복해서 찾는 것
- 세부 알고리즘 : 교안 6page

```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina' 
#한글설정 #rc는 plt 기본 속성들을 설정하는 함수
plt.rc('font',family='Malgun Gothic') #윈도우즈
plt.rc('axes', unicode_minus=False) #축설정 (축의 - 깨짐 방지)
```

```python
# 분류를 위한 가상의 데이터 생성
from sklearn.datasets import make_classification 
X, y = make_classification(n_samples=20,             # 데이터 갯수 (기본값-100)
                           n_features=2 ,            # 독립변수 갯수 (기본값-20) 
                           n_informative=2,          # 종속변수에 영향을 미치는 독립변수 갯수 
                           n_redundant=0,            # 종속변수에 영향을 미치지않는 독립변수 갯수
                           n_classes=2 ,             # 종속변수의 그룹 수(0,1)
                           n_clusters_per_class=1,   # 각 종속변수 그룹별 서브그룹 수 (1개라는건 서브그룹 없음)
                           random_state=123)
import numpy as np
np.c_[X,y]
```

```python
plt.scatter(x = X[:,0],
            y = X[:,1],
            c=y, s=100, edgecolors='k', lw=3)
plt.show()
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=2,       # 군의 갯수
               init="k-means++",   # k-means++(초기중심점을 서로 멀리 초기화)/ random(초기중심점랜덤초기화)
               n_init=10,          # 각 다른 초기값을 10번 실행해보고 가장 좋은 모델 선택
               max_iter=300)       # 각 실행에서 300번 반복(중심점 옮기기)

model.fit(X)              
```

```python
c0, c1 = model.cluster_centers_   # 모델이 정한 두개의 중심점
print(c0)
print(c1)
```

```python
print('modelLabel :',model.labels_)
pred = model.predict(X)
print("  예 측 값  :",pred)
print('  실 제 값  :', y)
```

```python
plt.scatter(x = X[:,0],
            y = X[:,1],
            c=pred, s=100, edgecolors='k', lw=3)
plt.scatter(x = c0[0], y = c0[1], s=200, marker='^', lw=5, c='g')
plt.scatter(x = c1[0], y = c1[1], s=200, marker='^', lw=5, c='g' )
plt.show()
```

# 3절. Hierarchical 클러스터링(계층적 군집)
- 군끼리 묶으면서 최종적으로 하나의 군집이 될 때까지 묶는 클러스터링
- 군집의 수를 미리 정해주지 않음
- 군집간의 거리 기반으로 클러스터링 

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head(2)
```

```python
# 계층적 군집 적용을 위해 라벨 인코딩
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
iris['species'] = le.fit_transform(iris.species)
```

```python
iris[::50]
```

```python
from scipy.cluster.hierarchy import linkage
# method : 거리계산을 어떻게 할 지
# method='complete' : 두 클러스터에서 가장 먼 거리를 이용
# method='single' : 두 클러스터에서 가장 먼 거리를 이용
# method='average' : 두 클러스터내의 각 점에서 다른 클러스터의 모든 점의 평균거리 이용 
cluster_model = linkage(iris, method='complete') 
```

```python
from scipy.cluster.hierarchy import dendrogram
dendrogram(cluster_model, labels=iris.index)
# plt.savefig('data/ch02_iris.png', dpi=300, bbox_inches="tight")
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster # 군집화
fcluster(cluster_model,
         t=4,  # dendrogram의 y축의 값
         criterion='distance' # 군집화 기준
        )
# criterion="distance"일 때, t값은 dendrogram의 y축 값
```

```python
fcluster(cluster_model, 
        t=3, 
        criterion='maxclust')
# criterion='maxclust' 일 때, t값은 클러스터의 갯수(군집화된 그룹 갯수)
```

```python
fcluster(cluster_model, 2, criterion='distance')
```

# 4절. DBSCAN 클러스터링
- 연결 기반(밀도기반) 군집화 알고리즘
- 단점 : 클러스터의 밀도가 일정하지 않으면 성능이 낮아짐 

```python
X1 = np.random.rand(200)*3 # 0<=X1<2의 실수 난수 
noise = np.random.normal(0,0.1,200)
Y1 = np.sin(X1) + noise
group1 = np.c_[X1,Y1, np.full(200,0)]
X2 = X1 + 1.5
Y2 = np.cos(X2) + noise
# plt.scatter(X2, Y2)
group2 = np.c_[X2, Y2, np.full(200,1)]
data = np.r_[group1,group2]

plt.scatter(data[:,0], data[:,1], c = data[:,2], s = 100, edgecolors='k', lw=2)
```

```python
data[:,:6]
```

```python
from sklearn.cluster import DBSCAN
# 반경 0.3 이내에 최소 10개는 있어야 군집 시작
db = DBSCAN(eps=0.3, min_samples=10)
db.fit(data[:,:2])
```

```python
labels = db.labels_ # 모델이 군집화한 결과 (-1은 이상치 인식)
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0) # 클러스터 수
n_noise_ = list(labels).count(-1)
print(f'이상치를 제외한 클러스터 수:{n_clusters_}, 이상치갯수:{n_noise_}개')
```

```python

plt.scatter(data[:,0], data[:,1], c = labels, s = 100, edgecolors='k', lw=2)
```

## 5.2 실루엣 계수
- 클러스터 개수, 소속을 모두 모르고 있을 경우 평가

```python
X1 = np.random.rand(200)*3 # 0<=X1<2의 실수 난수 
noise = np.random.normal(0,0.1,200)
Y1 = np.sin(X1) + noise
group1 = np.c_[X1,Y1, np.full(200,0)]
X2 = X1 + 1.5
Y2 = np.cos(X2) + noise
# plt.scatter(X2, Y2)
group2 = np.c_[X2, Y2, np.full(200,1)]
data = np.r_[group1,group2]

plt.scatter(data[:,0], data[:,1], c = data[:,2], s = 100, edgecolors='k', lw=2)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
range_n_clusters = [2,3,4,5,6,7,8,9,10]
for n_clusters in range_n_clusters:
    model = KMeans(n_clusters=n_clusters,
                   n_init=10,
                   random_state=2)
    model.fit(data[:,:-1])
    score = silhouette_score(X=data[:,:-1],
                             labels=model.labels_)
    print("클러스트 수 : {}, 실루엣 계수 : {}".format(n_clusters,score))
# 실루엣 계수가 가장 높은 6개로 클러스터링 할 예정
```

```python
model = KMeans(n_clusters=6, n_init=10, init="k-means++")
model.fit(data[:,:-1])
centers = model.cluster_centers_
colors = ['r','g','b','c','m','k'] # 4,2,3 vs 0,1,5
for i in range(len(colors)):
    plt.scatter(data[model.labels_==i,0], data[model.labels_==i,1], c=colors[i])
```
